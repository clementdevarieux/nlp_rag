```python

import os

from src.config import EXP_DATA
```

```python
json_path = []
for filename in os.listdir(EXP_DATA):
    print(filename)
    if ".json" in filename:
        json_path.append(EXP_DATA / filename)
```

```python
json_path
```

```python
import json


def extract_data_from_json_list(file_path):
    with open(file_path, 'r') as file:
        data_list = json.load(file)
        

    experiments_data = []
    for data in data_list:
        print(str(file_path))
        data_dict = {
            'rouge1': data['rouge1'][0],
            'rouge2': data['rouge2'][0],
            'rougeL': data['rougeL'][0],
            'rougeLsum': data['rougeLsum'][0],
            'isAccurateChunk': data['isAccurateChunk'],
            'experiment': "_".join(str(file_path).split('/')[-1].split('_')[2:-2]) # Extract experiment name from file name
        }
        experiments_data.append(data_dict)
        print(data_dict)
        

    return experiments_data

```

```python
data_list = []
for file_path in json_path:
    data_list.extend(extract_data_from_json_list(file_path))
```

```python
import pandas as pd

df = pd.DataFrame(data_list)
```

```python
df
```

```python
def reverse_experiment_name(value):
    vs : list = value.split('_')
    vs.reverse()
    return "_".join(vs)

df['experiment'] = df['experiment'].apply(reverse_experiment_name)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
```

```python
def plot_data(df):
    df = df.sort_values(by=['experiment'])
    plt.figure(figsize=(14, 8))
    df_melted = df.melt(id_vars=['experiment', 'isAccurateChunk'],
                    value_vars=['rouge1', 'rouge2', 'rougeL', 'rougeLsum'],
                    var_name='metric', value_name='score')


    sns.lineplot(data=df_melted, x='experiment', y='score', hue='metric', marker='o')
    plt.title('ROUGE Scores Across Experiments')
    plt.xlabel('Experiment')
    plt.ylabel('ROUGE Score')
    plt.legend(title='Metric')
    plt.xticks(rotation=90)
    plt.tight_layout()

    plt.show()
    
```

```python
plot_data(df)
```
